```python
import numpy as np
import matplotlib.pyplot as plt
import mkl
from IPython.display import Markdown, display

np.random.seed(1234)
mkl.set_num_threads(4)

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams["figure.figsize"] = [16, 9]
```

# Zadanie 1

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [3, 4]])
```

### Wymnóż macierze A i B. Wynik zapisz w macierzy W

```python
W = A @ B
```

### Test mnożenia macierzy

```python
if np.trace(W) == 45:
    display(Markdown('#### Prawidłowo!'))
else:
    display(Markdown('#### Błąd!'))
```

# Zadanie 2: zaimplementuj funkcję sigmoidalną

$\Large\sigma(x) = \frac{1}{1 + e^{-x}}$

```python
def sigma(x):
    wynik = 1/(1-np.exp(-x))
    return wynik
```

### Test funkcji sigmoidalnej

```python
z = np.random.uniform(size=100)*10.0 - 5.0
z = np.sort(z)

s = sigma(z)
plt.plot(z, s)
```
